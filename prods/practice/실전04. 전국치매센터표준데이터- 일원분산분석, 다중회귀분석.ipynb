```python
import pandas as pd
import numpy as np
```

```python
D = pd.read_csv('data/public_imbclty_cnter.csv',sep='\t')
D.shape
```

```python
D.head()
```

```python
"""
전처리

건축물면적 변수 결측치 제거(최종 행 수 236개)

소재지도로명주소를 통해 광역시와 지방행정구역(도) 정보 변수를 추가하라

이때 전라, 충청, 경상은 하나로 통합한다.

(ex:전라남도 -> 전라도, 충청남도 -> 충청도, 경상남도 -> 경상도)
"""
```

```python
D.dropna(subset=['bildngAr'], axis=0)
```

```python
D = D.dropna(subset=['bildngAr'], axis=0)
```

```python
search = ['전라남도','전라북도','경상남도','경상북도','충청남도','충청북도']
substitute = ['전라도','전라도','경상도','경상도','충청도','충청도']
D.rdnmadr.apply(lambda x : x.split(' ')[0]).replace(to_replace=search, value=substitute).value_counts()
```

```python
D['region'] = D.rdnmadr.apply(lambda x : x.split(' ')[0]).replace(to_replace=search, value=substitute).copy()
```

```python
D.head()
```

```python
# Q1
"""
Q1: 센터별 전체 임직원 수 변수(의사인원 + 간호사인원 + 사회복지사인원 + 

      기타  인원)를 추가하고 건축물면적(㎡)당 인원수를 확인하라

      이때 건축물면적(㎡)당 인원수 아웃라이어 갯수는?(tukey 방식 활용) 
"""
```

```python
D['allCo'] = D['doctrCo'] +  D['nurseCo'] + D['scrcsCo'] + D['etcHnfSttus']
D['dens'] =  D['allCo']/D['bildngAr']
D.head()
```

```python
D.dens.describe()
```

```python
# 이상치 확인
# The IQR (interquartile range) is the difference between Q3 and Q1.
Q1 = D.dens.describe()['25%']
Q3 = D.dens.describe()['75%']
IQR = Q3 - Q1
lb = Q1 - 1.5*IQR
hb = Q3 + 1.5*IQR
Q1, Q3, IQR, lb, hb
```

```python
Q1 = np.percentile(D.dens, 25)
Q3 = np.percentile(D.dens, 75)
IQR = Q3 - Q1
lb = Q1 - 1.5*IQR
hb = Q3 + 1.5*IQR
Q1, Q3, IQR, lb, hb
```

```python
D[(D.dens < lb)|(D.dens > hb)].shape
```

```python
D['outlier'] = np.where((D.dens > lb)&(D.dens < hb),  'normal', 'abnorbal')
D
```

```python
# Q2
_ = """
Q2: Q1에서 확인한 아웃라이어를 제거한 후 Q2, Q3 진행 

     도별 Q1에서 구한 건축물면적당 인원수 평균이 차이나는지 확인하고자 한다. 

     경기도, 충청도, 전라도, 경상도, 강원도 5개 도의 건축물면적당 인원수 평균이 

     통계적으로 유의미하게 차이가 난다고 볼 수 있는가?

     (등분산, 정규성을 가진다 가정하고 일원산분석 ANOVA 검정을 수행한다)

 

     차이가 있으면 Y, 없으면 N로 표시하고 

     차이가 있을 시 검정통계량을 소수점 둘째자리(셋째자리 버림)까지 표시하여라
"""
```

```python
D2 = D[(D.dens > lb)&(D.dens < hb)]
D2.shape
```

```python
D[D.region=='경기도'].shape, D[D.region=='충청도'].shape, D[D.region=='전라도'].shape, D[D.region=='경상도'].shape, D[D.region=='강원도'].shape, 
```

```python
# Q3
"""
Q3: 건축물면적, 총 임직원수, 의사수 3개 변수를 독립변수로 

     간호사수를 종속변수로 하여 다중 회귀분석을 실시하고자 한다.

     이때 전체 219개 센터를 대상으로 Split 함수를 사용하여 Train set과 

     Test set을 7:3으로 분할(seed=123)한 데이터로 모델을 설정하고자 한다.      

    TrainSet바탕의 모델을 TestSet으로 수행했을 때  간호사 인원수를 10명 이상 예측한 센터는 모두 몇개인가?
"""
```

```python
D2.head()[['bildngAr', 'allCo', 'doctrCo', 'nurseCo']]
```

```python
X = D2[['allCo', 'doctrCo','bildngAr']].values
```

```python
y = D2['nurseCo'].values
```

```python
from  sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.3, random_state=123)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
predict = model.predict(X_test)
predict
```

```python
pd.DataFrame(data={'predict':predict, 'true': y_test})
```

```python
model.score(X_train, y_train) # 결정계수
```

```python
model.coef_ # 회귀계수
```

```python
model.intercept_ # bias
```
