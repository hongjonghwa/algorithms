Q2

```
df3 = df2.loc[(df2['product']=='애완용품')&(df2['type']=='Online몰')]
df3.head()
```

```
df3['diff'] = df3.value.diff()
df3['prev'] = df3.value.shift()
```

```
df3.head()
```

```
df3['inc'] = df3.diff/df3.value
df3
```

```
df3.sort_values(by='inc', ascending=False)
```

Q3

```
df4 = df2.groupby(by='month').sum()
df4.head()
```

```
df4.shape
```

```
df4['pred'] = df4.value.ewm(min_periods=4, alpha=0.8).mean()
```

```
df4.tail()
```

```
pred = df4.loc[ 'month_2007','pred']
real = df4.loc['month_2007', 'value']
pred, real
```

```
#mape - mean absolute percent error
(pred-real)/real*100
```
