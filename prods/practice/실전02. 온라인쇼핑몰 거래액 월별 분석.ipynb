```python
# 다음 중 애완용품의 Online몰 거래액 중 지난달 대비 가장 높은 상승률을 보인 월은 언제인지 확인하시오.(소수 셋째자리까지 나타내시오)
```

```python
#df3 = df2.loc[(df2['product']=='애완용품')&(df2['type']=='Online몰')]
#df3.head()
df3 = df2[(df2['product']=='애완용품')&(df2['type']=='Online몰')][['mon', 'value']].set_index('mon')
```

```python
df3['prev'] = df3['value'].shift(1)
```

```python
df3
```

```python
df3['incr'] = df3['value'] / df3['prev']
```

```python
df3.sort_values(by='incr', ascending=False)
```

```python
# 월별 Online몰, On/OFFline몰 거래액 전체 합계를 구하고 지수 가중 이동평균 평활법(EWMA)을 사용하여
# 20년 7월 예측된 거래액 전체 합계와 실제 값을 비교하여 MAPE(평균절대백분율 오차)를 구하시오
#
# - 기간은 4, a값은 0.8을 사용할 것, 소수 셋째자리는 버리고 둘째자리까지 
```

```python
df4 = df2.groupby(by='mon').sum()
df4
```

```python
df4.shape
```

```python
df4['pred'] = df4['value'].ewm(min_periods=4, alpha=0.8).mean()
```

```python
df4.tail()
```

```python
pred = df4.loc[ 'month_2007','pred']
real = df4.loc['month_2007', 'value']
pred, real
```

```python
import numpy as np
#mape - mean absolute percent error
np.abs(pred-real)/real*100
```
