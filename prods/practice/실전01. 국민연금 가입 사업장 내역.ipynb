Q3 5월 기준 가입자수 기준으로 50위까지 확인하고, 50위까지의 사업장 대응표본 T검정을 통해 가입자수 평균 차이가 생겼는지 확인하라
(등분산, 정규성은 가지고 있다고 가정하고 진행한다.
 귀무가설을 기각 하는 지 여부와 함께 t-value 값을 소수 셋째자리 반올림, 둘째자리까지 나타내라)

```python
df[df.mon=='5월'].sort_values(by='total_cnt', ascending=False).head(50)
```

```python
merged = df[df.mon=='5월'].sort_values(by='total_cnt', ascending=False).head(50).join(
    df[df.mon=='6월'].set_index('company'),
    on="company",
    how="left",
    lsuffix="_before",
    rsuffix="_after"
)
merged
```

```python
merged[['company','total_cnt_before', 'total_cnt_after', 'mon_before', 'mon_after']]
```

```python
import scipy.stats as stats
```

```python
stats.ttest_rel(merged['total_cnt_before'], merged['total_cnt_after'])
```
