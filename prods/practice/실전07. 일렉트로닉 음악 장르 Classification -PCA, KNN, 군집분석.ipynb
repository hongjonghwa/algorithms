```python
# log 변환
D['ChromaVector1m_log'] = np.log(D.ChromaVector1m)
D['SpectralFluxstd_log'] = np.log(D.SpectralFluxstd)
D['ChromaVector1std_log'] = D.ChromaVector1std.map( lambda x: np.log(x)) # 이 방법도 동일함
```

```python
B = D.drop(['ChromaVector1m','SpectralFluxstd','ChromaVector1std'], axis=1)
B.head()
```

```python
# Q1
_ = """
Q1: 먼저 18개 변수를 차원 축소를 통해 변수의 수를 줄이고자 한다. 
    주성분분석 PCA를 진행하고자 하는데 이를 위해 표준화를 진행한다. 
    이때 표준화는 standardscaler 방식으로 따른다. 
    다음 표준화를 진행한 변수를 바탕으로 PCA 분석을 수행한다. 
    (SVD solver:auto, Seed:13, 나머지 디폴트) 
    이때 다섯번째 주성분의 누적 분산 비율을 구하시오.
    (소수 셋째자리 버림 둘째자리 구하시오) 
"""
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scaler = StandardScaler()
```

```python
pca = PCA(n_components=5, svd_solver="auto", random_state=13)
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([('scaler', scaler),
                 ('pca', pca)])
```

```python
pipe.fit(B.drop('class', axis=1))
```

```python
pca.components_
```

```python
pca.get_covariance()
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
np.sum(pca.explained_variance_ratio_[0:5])
```

```python
# Q2 
_ = """
Q2: 23개 장르를 보다 압축하기 위해 계층적 군집(Hierarchical clustering) 모형을 수행하고자 한다. 
    군집 간거리계산 방법은 유클리드 거리를 기록하고 병합은 와드 연결법으로 진행한다.
    (나머지 디폴트) 
    이때 변수는 Q1을 통해 수행한 0부터 4까지 다섯개까지 합성변수를 바탕으로 하고 군집간 거리를 통해 3개의 군집으로 나누어 본다. 
    이때 한 군집으로의 쏠림현상이 가장 큰 장르는 무엇인지 서술하시오.
"""
```

```python
principals = pipe.transform(B.drop('class', axis=1))
```

```python
pd.DataFrame(data=principals)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
aggcluster = AgglomerativeClustering(n_clusters=3, linkage="ward", affinity="euclidean")
```

```python
aggcluster.fit(principals)
```

```python
aggcluster.labels_
```

```python
pd.crosstab(aggcluster.labels_, B['class'], margins=True)
```

```python
pd.crosstab(aggcluster.labels_, B['class'], margins=True).T
```

```python
# Q3
_ = """
Q3: Q1-Q2를 종합적으로 KNN 분류 예측 모형을 만들고자 한다. 
먼저 Q2에서 군집분석 수행한 결과를 바탕으로 
데이터 분할을 트레인셋: 테스트셋 7:3 비율로 진행한다.(Seed:13)
다음 독립변수로는 앞서진행한 다섯개 주성분 합성변수로 진행하고 
종속변수는 3개의 클러스터 값으로 구분된 군집정보가 들어있는 변수를 지정한다.
이때 number of neighbors K값은 9로 지정하고 거리측정 방식은 맨하탄 거리측정으로 진행한다.
이후 KNN분류 예측 모형 평가를 진행하고자 한다. 
테스트 셋 690개에 대해 House 장르가 많은 비중을 차지했던 클러스터 집단에 해당하는 라벨값 Precision을 구하시오
(소수 셋째자리 버림 둘째자리 구하시오) 
"""
```

```python
# house 는 클러스터 0
pd.DataFrame({'cluster':aggcluster.labels_}).head()
```

```python
pd.DataFrame(data=principals, columns=["pca1","pca2","pca3","pca4","pca5",]).head()
```

```python
X = pd.DataFrame(data=principals, columns=["pca1","pca2","pca3","pca4","pca5",])
y = aggcluster.labels_
```

```python
X.shape, y.shape
```

```python

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size = 0.3,random_state = 13)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=9, p=1)
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
pd.DataFrame({"true":y_test, "predict": prediction}).head()
```

```python
from sklearn.metrics import recall_score, precision_score
```

```python
precision_score(y_test == 0, prediction == 0)
```

```python
recall_score(y_test == 0, prediction == 0)
```

```python
pd.crosstab(y_test, prediction)
```
