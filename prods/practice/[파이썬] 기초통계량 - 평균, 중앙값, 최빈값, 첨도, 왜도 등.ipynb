```python
df2.loc[:, 'target'] = wines.target.astype(str).value
```

```python
df2
```

```python
df2.apply('mode').T
```

```python
import  scipy.stats
```

```python
df2.apply(scipy.stats.mode).T
```

```python
df_mode = pd.DataFrame(df2.apply(scipy.stats.mode)).T.drop('target', axis=1)
df_mode.index = ['mode']
df_null = pd.DataFrame(df2.isnull().sum()).T.drop('target', axis=1)
df_null.index = ['null_count']
df_skew = df2.drop('target', axis=1).apply(['skew', 'kurtosis'])
df_mode, df_null, df_skew
```

```python
pd.concat( [df2.describe(),df_mode, df_null, df_skew])
```

```python
df2.target.value_counts()

```

```python
df2['target2'] = np.where(df2['proline'] % 2 == 0, '짝', '홀')
df2
```

```python
pd.crosstab(index=df2['target'], columns=df2['target2'])
```
