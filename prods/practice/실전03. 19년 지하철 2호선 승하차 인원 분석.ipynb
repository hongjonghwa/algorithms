```python
#Q3
"""
2호선 전체 역을 행으로 1부터 12월까지 월을 열으로 승하차 총 인원을
바탕으로 데이터를 나열한 후 Minmax- normalization을 수행한다. 
이후 K-means 군집 분석을 통하여 1월부터 12월까지 normalization한 
12개의 변수를 바탕으로 4개의 군집으로 분류한다. 
강남역과 동일한 군집을 가진 군집들의 1월 총 승하차 평균 승객은? 
(seed는 1234로 셋팅, 강남역도 포함하여 계산 해 정수 값만 나타낼 것)
"""
```

```python
D.head()
```

```python
D['ymon'] = D.date.str[0:7]
D
```

```python
D3 = D.pivot_table(index='station', columns='ymon', values='cnt',  aggfunc='sum')
D3.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline
```

```python
model = KMeans(n_clusters=4, random_state=1234)
pipeline = Pipeline(steps=[('minmaxscaler',MinMaxScaler()),('kmeans', model)])
```

```python
pipeline.fit(D3)
```

```python
model.labels_
```

```python
pipeline.predict(D3)
```
