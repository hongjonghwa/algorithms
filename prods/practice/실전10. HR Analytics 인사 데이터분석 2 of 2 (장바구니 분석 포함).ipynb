```python
# Q2
_ = """
Q2: HR 데이터들을 바탕으로 임직원을 유사한 집단으로 클러스터링하고 퇴직률이 높은 집단을 확인하고자 한다.
   방식은 계층적 군집 분석을 사용하며 이를 위해 변수 추가 등 전처리 과정을 진행할 계획이다. 


1.변수 추가(변수명: top_department_Y) 
  퇴직률이 높은 부서(5곳)에 해당하는지 여부 
  'hr', 'accounting', 'technical', 'support', 'sales' 맞으면 1, 아닐 경우 0 

2.급여 수준 더미변수화 
  (범주 수 - 1 적용, medium 범주 변수 제외) 

3.데이터 정규화(5개 변수 Minmax 방식으로 정규화 시행) 
  (satisfaction_level, last_evaluation, number_of_projects, average_monthly_hours, years_at_company)



위 전처리 과정 후 계층적 군집 분석 수행(Hierarchical Clustering), 
전처리 통해 나온 8개 변수 + 2개 변수 추가(work_accident, promotion_last_5years) 총 10개 변수를 활용한다. 

(유클리드 거리 계산, 와드 연결법 진행)

​

군집분석 수행 후 14,999명 임직원 DB를 4개의 군집으로 분류할 때 임직원 퇴직률
( 군집에 해당하는 임직원 중 퇴사한 임직원 ÷ 군집에 해당하는 임직원 수)이 가장 낮은 군집의 퇴직률을 구하시오
"""
```

```python
# Q3
_ = """
Q3: 퇴직에 영향을 끼치는 요인들을 따져보기 위해 연관성 분석을 수행하고자 한다. 

먼저 아래 2가지 전처리를 수행한다. 

​

1. 수치형 변수 satisfaction_level, last_evaluation를 각각 등급별로 나눠 명목변수로 변환하고자 한다.

2개 변수 각각 1분위수(1Q) 이하일 경우 '하', 3분위수(3Q) 초과일 경우 '상' 그 사이는 

'중'으로 변환한 변수를 추가한다. 

(변수명 : satisfaction_level_type, last_evaluation_type)

​

2. 월 평균 근무 시간과 관련한 명목변수를 만든다.

월 평균 근무 시간이 200을 초과하면 'Y', 아닐 경우 'N'으로 기록한다.

(변수명 : average_monthly_hours_over200)

​

전처리를 통해 새롭게 만든 3개 명목변수와 

number_of_projects, years_at_company, work_accident, left, promotion_last_5years,

department, salary - 7개 변수를 바탕으로 연관성 분성을 수행한다.(Association Rule) 

(3가지 조건 외 디폴트 적용, Min Supprot - 0.05, Min Confidence - 0.8, Min Lift - 1) 

​

선행은 변수 2개 이하이며 후행은 단일 조건일 때 규칙은 총 380개 나온다. 

이때 퇴사를 하는 경우('left: 1')에 해당하는 규칙 중 lift값이 가장 높은 규칙은?

​

ex) number_of_projects: 2, last_evaluation_type: 하
"""
```

```python
D.describe()
```

```python
Q1 = D.satisfaction_level.quantile(.25)
Q3 = D.satisfaction_level.quantile(.75)
Q1, Q3
```

```python
pd.DataFrame(np.where (D.satisfaction_level<=Q1 , '하',  np.where (D.satisfaction_level>Q3 , '상', '중')   ), columns=['x']).value_counts()
```

```python
D['satisfaction_level_type'] = np.where (D.satisfaction_level<=Q1 , '하',  np.where (D.satisfaction_level>Q3 , '상', '중')   )
D
```

```python
Q1 = D.last_evaluation.quantile(.25)
Q3 = D.last_evaluation.quantile(.75)
```

```python
D['last_evaluation_type'] = np.where (D.last_evaluation<=Q1 , '하',  np.where (D.last_evaluation > Q3 , '상', '중') )
D
```

```python
D['average_monthly_hours_over200'] = np.where(D['average_monthly_hours'] > 200 , 'Y', 'N')
D
```

```python
df3 = D[['satisfaction_level_type','last_evaluation_type','average_monthly_hours_over200',
      'number_of_projects', 'years_at_company', 'work_accident', 'left', 'promotion_last_5years', 'department', 'salary'
      ]]
df3
```

```python
df3.info()
```

```python
df3.number_of_projects = df3.number_of_projects.astype(str).copy()
df3.years_at_company = df3.years_at_company.astype(str).copy()
df3.work_accident = df3.work_accident.astype(str).copy()
df3.left = df3.left.astype(str).copy()
df3.promotion_last_5years = df3.promotion_last_5years.astype(str).copy()
df3.info()
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
itemset = apriori(pd.get_dummies(df3), use_colnames=True, min_support= 0.05 )
itemset
```

```python
df3_rules = association_rules(df = itemset, metric="confidence", min_threshold=0.8)
df3_rules
```

```python
df3_rules['antecedents_count'] = df3_rules.antecedents.map(lambda x :len(frozenset(x)))
```

```python
df3_rules['consequents_count'] = df3_rules.consequents.map(lambda x :len(x))
```

```python
df3_rules[
    (df3_rules.antecedents_count <= 2) & 
    (df3_rules.consequents_count == 1) & 
    (df3_rules.consequents == frozenset({'left_1'})) ]
```
