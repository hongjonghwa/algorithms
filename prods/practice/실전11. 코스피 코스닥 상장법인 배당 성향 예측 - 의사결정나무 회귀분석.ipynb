```python
import numpy as np
import pandas as pd
```

```python
d1 = pd.read_csv('data/company_info_19.csv')
d2 = pd.read_csv('data/company_financial_19.csv')
d3 = pd.read_csv('data/company_dividen_19.csv')
```

```python
D = pd.merge(pd.merge(d1, d2, on="com_code"), d3, on="com_code")
D
```

```python
D['devidend_ratio'] = np.round(D.com_totaldevidend_19 / D.com_netincome_19, decimals=3)
D
```

```python
D = D[ D.devidend_ratio > 0 ]
D
```

```python
D.describe()
```

```python
Q1 = np.quantile(D.devidend_ratio, 0.25)
Q3 = np.quantile(D.devidend_ratio, 0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
Q1, Q3, IQR, lower, upper
```

```python
D = D[ (D.devidend_ratio > lower) & (D.devidend_ratio < upper) ]
```

```python
# Q2
_ = """
문제2) 각종 기업분석 안전성 지표들을 산출해 내고 
       이것이 배당성향에 얼마만큼 영향을 미치는지 상관관계를 살펴보고자 한다. 
       문제1번아래 6가지 지표를 파생변수로 만들고 이 6가지 지표 중 
       문제 1번을 통해 생성한 배당성향과의 상관계수 절대값이 가장 높은 변수는 무엇인지 쓰고 
       상관계수 절대값을 확인하라(피어슨 방식 활용, 소수점 셋째자리에서 버림, 둘째자리까지 나타내시오)

변수명 정의

Operating_margin 영업이익률: 영업이익/매출액
Netprofit_margin 순이익률: 당기순이익/매출액
Dept_ratio 부채비율: 부채총계/자본총계
Banker_ratio 유동비율: 유동자산/유동부채
Quick_ratio 당좌비율: ( 유동자산 - 재고자산 ) / 유동부채
Reserve_ration 유보율: (자본잉여금 + 이익잉여금) / (납입)자본금

"""
```

```python
D['Operating_margin'] = D['OperatingIncomeLoss_19'] / D['Revenue_19']
D['Netprofit_margin'] = D['com_netincome_19'] / D['Revenue_19']
D['Dept_ratio'] = D['Liabilities_19'] / D['Equity_19']
D['Banker_ratio'] = D['CurrentAssets_19'] / D['CurrentLiabilities_19']
D['Quick_ratio'] = (D['CurrentAssets_19']-D['Inventories_19'] )/ D['CurrentLiabilities_19']
D['Reserve_ration'] = (D['CapitalSurplus_19']+D['RetainedEarnings_19']) / D['IssuedCapital_19']
```

```python
D[['devidend_ratio',
   'Operating_margin','Netprofit_margin','Dept_ratio','Banker_ratio','Quick_ratio','Reserve_ration'
  ]]
```

```python
D[['devidend_ratio',
   'Operating_margin','Netprofit_margin','Dept_ratio','Banker_ratio','Quick_ratio','Reserve_ration'
  ]].corr(method='pearson')
```

```python
# Q3
_ = """
문제3) 배당성향을 예측하는 모델을 만들고자 한다. 
       의사결정나무(Decision Tree)를 활용한 회귀분석을 하고자 한다.
       먼저 트레인셋, 테스트 셋을 7:3 으로 분할하며
       (sklearn.model_selection 라이브러리 train_test_split 함수 활용, Seed: 1234) 
       트레인셋을 바탕으로 아래 변수들을 바탕으로 학습해 테스트셋 종속변수를 예측하고자 한다. 
       다음 문제 2에서 추가한 변수 6개를 바탕으로 독립변수로 지정하고 배당성향을 종속변수로 한다. 

독립변수: 영업이익률, 순이익률, 부채비율, 유동비율, 당좌비율, 유보율
종속변수: 배당성향
옵션값: Criterion-MSE, Splitter-Best, Max Depth- 7, Max Features - 5, Seed - 1234 나머지 디폴트
        (sklearn.tree 라이브러리에서 DecisionTreeRegressor 함수를 활용할 것) 
        

테스트셋 78개에 대해 예측한 배당성향과 실제 배당성향 유클리디언 거리를 각각 측정하고 평균을 내시오.
(소수 셋째자리에서 버리고 둘째자리까지 나타내시오) 

"""
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    D[['Operating_margin','Netprofit_margin','Dept_ratio','Banker_ratio','Quick_ratio','Reserve_ration']],
    D['devidend_ratio'],
    test_size=0.3, random_state=1234)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
model = DecisionTreeRegressor(criterion="mse",splitter='best', max_depth=7, max_features=5, random_state=1234)
```

```python
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
pred
```
